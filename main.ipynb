# **AI Resume Screening System with Tracing**


**STEP 1: Add Job Description + Resumes**

```python
# Job Description
job_description = """
We are looking for a Data Scientist with:
- Python, Machine Learning, Deep Learning
- Experience with Pandas, NumPy, Scikit-learn
- Knowledge of NLP is a plus
- 2+ years experience preferred
"""

# Strong Resume
resume_strong = """
John Doe
Skills: Python, Machine Learning, Deep Learning, NLP, Pandas, NumPy, Scikit-learn
Experience: 3 years as Data Scientist
Projects: Built ML models for prediction and NLP applications
"""

# Average Resume
resume_avg = """
Jane Smith
Skills: Python, Machine Learning, Pandas
Experience: 1 year as Data Analyst
Projects: Basic ML models
"""

# Weak Resume
resume_weak = """
Sam Brown
Skills: Excel, Communication
Experience: Fresher
Projects: None
"""
```

**STEP 2: Skill Extraction Function**

```python
def extract_skills(text):
    skills_list = ["Python", "Machine Learning", "Deep Learning", "NLP", "Pandas", "NumPy", "Scikit-learn"]

    extracted = []
    for skill in skills_list:
        if skill.lower() in text.lower():
            extracted.append(skill)

    return extracted
```

**STEP 3: Matching Logic**

```python
def match_skills(resume_skills, jd_skills):
    matched = set(resume_skills).intersection(set(jd_skills))
    return list(matched)
```

**STEP 4: Scoring**

```python
def calculate_score(matched, total_required):
    return int((len(matched) / total_required) * 100)
```

**STEP 5: Explanation**

```python
def generate_explanation(score, matched, missing):
    return f"""
Score: {score}/100

Matched Skills: {matched}
Missing Skills: {missing}

Explanation:
The candidate matches {len(matched)} required skills.
Missing important skills: {missing}.
Overall suitability is {'High' if score > 70 else 'Moderate' if score > 40 else 'Low'}.
"""
```

**STEP 6: Full Pipeline**

```python
def pipeline(resume):
    jd_skills = extract_skills(job_description)
    resume_skills = extract_skills(resume)

    matched = match_skills(resume_skills, jd_skills)
    missing = list(set(jd_skills) - set(matched))

    score = calculate_score(matched, len(jd_skills))

    explanation = generate_explanation(score, matched, missing)

    return explanation
```

**STEP 7: Run All 3 Cases**

```python
print("STRONG CANDIDATE:\n", pipeline(resume_strong))
print("\n----------------------\n")

print("AVERAGE CANDIDATE:\n", pipeline(resume_avg))
print("\n----------------------\n")

print("WEAK CANDIDATE:\n", pipeline(resume_weak))
```
